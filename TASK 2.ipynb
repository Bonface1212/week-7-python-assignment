```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load Iris dataset and convert to DataFrame
iris_data = load_iris(as_frame=True)
iris = iris_data.frame
iris['species'] = iris_data.target_names[iris_data.target]
iris = iris[[*iris_data.feature_names, 'species']]
iris.head()
```

```python
# Basic statistics
iris.describe()
```

```python
# Median and standard deviation
median_values = iris.median(numeric_only=True)
std_devs = iris.std(numeric_only=True)

print("Median Values:")
print(median_values)

print("\nStandard Deviations:")
print(std_devs)
```

```python
# Group by species and compute the mean
grouped_means = iris.groupby('species').mean(numeric_only=True)
grouped_means
```

```python
### 🔍 Observations:
- **Setosa** has the shortest petals (~1.46 cm on average).
- **Virginica** has the longest petals (~5.55 cm).
- **Versicolor** has the narrowest sepal width (~2.77 cm).
- **Petal length** shows a strong pattern to differentiate flower species.
```
