```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris_data = load_iris(as_frame=True)
iris_df = iris_data.frame
iris_df['species'] = iris_data.target_names[iris_data.target]
iris_df = iris_df[[*iris_data.feature_names, 'species']]

# Create a fake 'date' column to simulate time-series data
np.random.seed(0)
iris_df['date'] = pd.date_range(start='2023-01-01', periods=len(iris_df), freq='D')
iris_df['daily_measure'] = iris_df['sepal length (cm)'] + np.random.normal(0, 0.2, size=len(iris_df))
```

```python
plt.figure(figsize=(10, 5))
plt.plot(iris_df['date'], iris_df['daily_measure'], color='green')
plt.title('Line Chart: Simulated Sepal Length Over Time')
plt.xlabel('Date')
plt.ylabel('Sepal Length + Random Variation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_petal_length = iris_df.groupby('species')['petal length (cm)'].mean()

plt.figure(figsize=(7, 5))
avg_petal_length.plot(kind='bar', color=['red', 'blue', 'purple'])
plt.title('Bar Chart: Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.hist(iris_df['sepal width (cm)'], bins=15, color='orange', edgecolor='black')
plt.title('Histogram: Sepal Width Distribution')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(
    data=iris_df,
    x='sepal length (cm)',
    y='petal length (cm)',
    hue='species',
    palette='Set1'
)
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```
